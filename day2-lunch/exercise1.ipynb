```python
fs = open('fly.txt', 'r')
```

```python
fs2 = open('newfly.txt', 'w')
```

```python
for line in fs:
    if 'DROME' in line:
        fields = line.rstrip('\n\r').split()
        flybase = fields[-1]
        AC = fields[-2]
        newline = flybase + "\t"  + AC #string together flybase and AC to be able to add to newfly.txt simultaneously
        fs2.write(newline)
        print(flybase , AC)
```
