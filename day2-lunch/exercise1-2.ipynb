```python
interest = "DROME"

#have a gene id, want to match it to a uniprot id
#id_dict = {"gene1" : "uniprot1","gene2" : "uniprot2"}

id_dict = {}
for line in fs:
        if "DROME" not in line:
            continue
        else:
            if "FBgn" in line:
                fields = line.split()
                gene_id = fields[-1]
                uniprot_id = fields[-2]
            
            id_dict[gene_id] = uniprot_id #assign the keys and value to our empty dictionary

id_dict
#create dictionary outside of for loop bc everytime you go through loop it would rewrite th edictionary
```

```python
#go through lines in ctab file
# for every line, replace  gene_id field with uniprot_id
#gett uniprot_id from id_dict
#print line of ctab fine w  uniprtot id in place of gene id

for line in ctab:
    fields = line.rstrip("\n\r").split()
    gene_id = fields[8]
    
    #if gene_id is not in id_dict:
    if id_dict.get(gene_id, "not here") == "not here":    
        continue
    else:
        uniprot_id = id_dict[gene_id]
        
        first_half = fields[0:8]
        second_half = fields[9:]
        
        first_half_string = "\t".join(first_half)
        second_half_string = "\t".join(second_half)
        
        print(first_half_string + "\t" + uniprot_id + "\t" + second_half_string)
```

```python
 def replace_geneid(some_dict, some_ctab_file, some_arg3):
        
        #arg3 tells you waht to do if gene id is not in dictionary
        #option 1: you skip line
        #option 2: you print line, replace gene_id with some other thing ("NA")
    count = 0
    for line in some_ctab_file:
        count += 1
        if count < 101:
            fields = line.rstrip("\n\r").split()
            gene_id = fields[8]
            first_half = fields[1:8]
            second_half = fields[9:]
            first_half_string = "\t".join(first_half)
            second_half_string = "\t".join(second_half)

            if some_arg3 == "skip":

                #figure out what uniprot id is and print
                #if gene_id is not in id_dict:
                if some_dict.get(gene_id) == None:    
                    continue
                else:
                    uniprot_id = some_dict.get(gene_id)

                    print(first_half_string + "\t" + uniprot_id + "\t" + second_half_string)
            else:
            #print the line, replace gene_id with some other thing

                    if some_dict.get(gene_id, "not here") == "not here":    
                        #replace gene id with something else

                        print(first_half_string + "\t" + "not here" + "\t" + second_half_string)

                    else:
                        uniprot_id = some_dict.get(gene_id)

                        print(first_half_string + "\t" + uniprot_id + "\t" + second_half_string)
```
