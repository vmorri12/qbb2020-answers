```python
import numpy
DF_list = []
GQ_list = []
AF_list = []

for line in open('/Users/cmdb/qbb2020-answers/quantbio/week2/step7output.txt'):
    if line.startswith('#'):
        continue
    fields = line.strip('\r\n').split('\t')
    subfields = fields[7].split(';') #info fields
    subfields2 = fields[9].split(':') #format fields
    subfields3 = subfields[3]
    DF = subfields2[2]
    DF_list.append(DF)
    GQ = subfields2[1]
    GQ_list.append(GQ)
    AF = subfields3[3:]
    AF_list.append(AF)
    
print(DF_list)
    
```

```python
print(GQ_list)
```

```python
print(AF_list)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.hist(DF_list, bins = 30)
ax.set_xlabel("Depth")
ax.set_ylabel("Varient Genotypes")
plt.title("Depth Distribution Across Varient Genotypes in Yeast")

plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(GQ_list, bins =30)
ax.set_xlabel("Quality")
ax.set_ylabel("Varient Genotypes")
plt.title("Quality Distribution Across Varient Genotypes in Yeast")

plt.show()

```

```python
fig, ax = plt.subplots()
ax.hist(AF_list, bins =30)
ax.set_xlabel("Allele Frequency")
ax.set_ylabel("Varient Genotypes")
plt.title("Allele Frequency Across Varient Genotypes in Yeast")

plt.show()
```

```python
fig, ax = plt.subplots(ncols=3)
ax[0].hist(DF_list, bins = 30)
ax[0].set_xlabel("Depth")
ax[0].set_ylabel("Varient Genotypes")
ax[0].set_title("Depth Distribution")
ax[1].hist(GQ_list, bins =30)
ax[1].set_xlabel("Quality")
ax[1].set_ylabel("Varient Genotypes")
ax[1].set_title("Quality Distribution")
ax[2].hist(AF_list, bins =30)
ax[2].set_xlabel("Allele Frequency")
ax[2].set_ylabel("Varient Genotypes")
ax[2].set_title("Allele Frequency Spectrum")

plt.tight_layout()
```
