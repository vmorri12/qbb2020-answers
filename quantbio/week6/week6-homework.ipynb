```python
#compare e4 and e5 files 
#remove sites that have zero from e4
methyl_e4 = np.genfromtxt("/Users/cmdb/qbb2020-answers/quantbio/week6/SRR3083926_1.chr6_bismark_bt2_pe.sorted.bedGraph", usecols = (1, 3), skip_header = 1) #assumes number is float  
e4_no_zero = []
for i in range(len(methyl_e4)):
    if (methyl_e4[i, 1]) == 0:
        continue
    else:
         e4_no_zero.append([int(methyl_e4[i, 0]), methyl_e4[i, 1]])
        
print(e4_no_zero)
```

```python
#remove sites that have zero from e5
methyl_e5 = np.genfromtxt("/Users/cmdb/qbb2020-answers/quantbio/week6/SRR3083929_1.chr6_bismark_bt2_pe.sorted.bedGraph", usecols = (1, 3), skip_header = 1)
e5_no_zero = []
for i in range(len(methyl_e5)):
    if (methyl_e5[i, 1]) == 0:
        continue
    else:
        e5_no_zero.append([int(methyl_e5[i, 0]), methyl_e5[i, 1]])
        
print(e5_no_zero)
```

```python
#read in e5 bedgraph 
#use while bc if use for loop would not be able to alternate increment
e4_count = 0
e5_count = 0
fold_change = []

while(e4_count < len(e4_no_zero) and e5_count < len(e5_no_zero)):
    if e4_no_zero[e4_count][0] == e5_no_zero[e5_count][0]:
        fold_change.append([e4_no_zero[e4_count][0], e5_no_zero[e5_count][1], e4_no_zero[e4_count][1]])
        e4_count +=1
        e5_count +=1
    elif  e4_no_zero[e4_count][0] < e5_no_zero[e5_count][0]:
        e4_count +=1
    else:
        e5_count += 1
print(fold_change)
```

```python
#make list of genes in mm10 reference sequence 
no_reps = []
gene_array = np.genfromtxt("/Users/cmdb/qbb2020-answers/quantbio/week6/mm10_refseq_genes_chr6_50M_60M.bed", usecols = (4,5,12), dtype = str)



for i in range(len(gene_array)):
    if gene_array[1][2] in no_reps:
        continue
    else:
        no_reps.append([gene_array[i][2], int(gene_array[i][0])])
print(no_reps)
```

```python
#did not want to eliminate replicates, so got around this by making a dictionary of each gene with a value of a list for the postion of each methylation site within
gene_dict = {}
for i in range(len(gene_array)):
    gene_dict.setdefault(gene_array[i][2], [])
    gene_dict[gene_array[i][2]].append((int(gene_array[i][0]), int(gene_array[i][1])))

print(gene_dict)
```

```python
gene_fold_change = {}
for key, value in gene_dict.items():
    gene_fold_change.setdefault(key, [0.0, 0.0, 0.0])
    for i in range(0, len(fold_change)):
        for j in value:
            if j[0] <= fold_change[i][0] and fold_change[i][0] <= j[1]:
                gene_fold_change[key][0] += fold_change[i][1]
                gene_fold_change[key][1] += fold_change[i][2]
                gene_fold_change[key][2] += 1
                break 
    if gene_fold_change[key][2] == 0:
        del gene_fold_change[key]    #remove zeros from dictionary
                                
            
                                

print(gene_fold_change)
```

```python
fold_change_output = []
for key, value in gene_fold_change.items():
    fold_change_output.append([key, (value[0]/value[2])/(value[1]/value[2])])#calculate average
    
print(fold_change_output)
```

```python
np.savetxt("methylation_enrichment.txt", fold_change_output, fmt = "%s", delimiter="\t")
```
